Task: From the given ‘Iris’ dataset, predict the optimum number of clusters and represent it visually. 


```python
#importing required libraries & dataset for the project
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
from sklearn import datasets
%matplotlib inline

#reading the dataset 
data = datasets.load_iris() 
data
df=data
```

```python
# Finding the optimum number of clusters for k-means classification

x=pd.DataFrame(data.data)
x.columns=['Sepal_Length','Sepal_width','Petal_Length','Petal_width']

from sklearn.cluster import KMeans
wcss = []

#A simple method for deciding the optimum value is by trying different values of ‘K’.
#For this, we can simply make use of a ‘for’ loop. 

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', 
                    max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
    
# Plotting the results onto a line graph, 
# allowing us to observe 'The elbow'
plt.plot(range(1, 11), wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
x=pd.DataFrame(data.data)
x.columns=['Sepal_Length','Sepal_width','Petal_Length','Petal_width']

y=pd.DataFrame(data.target)
y.columns=['Targets']
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(x)
```

```python
plt.scatter(x.Sepal_Length, x.Sepal_width)
```

```python
colormap=np.array(['Red','green','blue'])
plt.scatter(x.Sepal_Length, x.Sepal_width,c=colormap[y.Targets],s=40)
plt.title('Actual Cluster')
```

```python
colormap=np.array(['blue','red','green'])
plt.scatter(x.Sepal_Length, x.Sepal_width,c=colormap[kmeans.labels_],s=40)
plt.title('After K-means Clustering ')
```
